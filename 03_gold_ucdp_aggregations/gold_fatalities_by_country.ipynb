```python
df_silver = spark.table(f'{catalog}.{silver_schema}.events_cleaned')
```

```python
gold_fatalities_by_country = df_silver \
    .groupBy("country", "year") \
    .agg(
        F.sum("best_estimate_deaths").alias("total_deaths"),
        F.sum("deaths_civilians").alias("civilian_deaths"),
        F.sum("deaths_a").alias("side_a_deaths"),
        F.sum("deaths_b").alias("side_b_deaths"),
        F.sum("deaths_unknown").alias("unknown_deaths")
    ) \
    .orderBy("country", "year")

gold_fatalities_by_country.write \
    .format("delta") \
    .mode("overwrite") \
    .option('overwriteSchema', True)\
    .saveAsTable("west_africa_conflict.gold.fatalities_by_country")

print(" gold.fatalities_by_country written")
display(gold_fatalities_by_country)
```
