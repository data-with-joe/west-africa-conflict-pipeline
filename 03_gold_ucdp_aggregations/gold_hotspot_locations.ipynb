```python
df_silver = spark.table(f'{catalog}.{silver_schema}.events_cleaned')
```

```python
gold_hotspot_locations = df_silver \
    .groupBy("country", "state_province") \
    .agg(
        F.count("event_id").alias("total_incidents"),
        F.sum("best_estimate_deaths").alias("total_deaths"),
        F.sum("deaths_civilians").alias("civilian_deaths")
    ) \
    .orderBy(F.desc("total_incidents"))

gold_hotspot_locations.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("west_africa_conflict.gold.hotspot_locations")

print(" gold.hotspot_locations written")
display(gold_hotspot_locations)
```
