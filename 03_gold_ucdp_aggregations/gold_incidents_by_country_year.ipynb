```python
df_silver = spark.table(f"{catalog}.{silver_schema}.events_cleaned")
```

```python
print(f'row count: {df_silver.count()}')
```

```python
gold_incidents_by_country_year  =(df_silver
    .groupBy('country', 'year')
    .agg(F.count('event_id').alias('total_incidents'))
    .orderBy('country', 'year')
    
    )
```

```python
(
    gold_incidents_by_country_year 
        .write
        .format('delta')
        .mode('overwrite')
        .saveAsTable(f'{catalog}.{gold_schema}.incidents_by_country_year')
)
```
