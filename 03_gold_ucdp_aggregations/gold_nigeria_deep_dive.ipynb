```python
df_silver = spark.table(f'{catalog}.{silver_schema}.events_cleaned')
```

```python
gold_nigeria_deep_dive = df_silver \
    .filter(df_silver["country"] == "Nigeria") \
    .groupBy("year", "violence_type_desc") \
    .agg(
        F.count("event_id").alias("total_incidents"),
        F.sum("best_estimate_deaths").alias("total_deaths"),
        F.sum("deaths_civilians").alias("civilian_deaths")
    ) \
    .orderBy("year")

gold_nigeria_deep_dive.write \
    .format("delta") \
    .mode("overwrite") \
    .option("overwriteSchema", "true") \
    .saveAsTable("west_africa_conflict.gold.nigeria_deep_dive")

print("gold.nigeria_deep_dive written")
display(gold_nigeria_deep_dive)
```
