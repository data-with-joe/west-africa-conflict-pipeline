```python
df_silver =spark.table( f'{catalog}.{silver_schema}.events_cleaned')
```

```python


gold_violence_type_breakdown = df_silver \
    .groupBy("country", "violence_type_desc") \
    .agg(F.count("event_id").alias("total_incidents")) \
    .orderBy("country", F.desc("total_incidents"))

gold_violence_type_breakdown.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable("west_africa_conflict.gold.violence_type_breakdown")

print(" gold.violence_type_breakdown written")
display(gold_violence_type_breakdown)
```
